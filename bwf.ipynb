```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import re
import requests
```

```python
target = 'http://bwfbadminton.com/rankings/2/bwf-world-rankings/6/men-s-singles/{year}/{week}?rows_per_page=5000'
xd_test = 'http://bwfbadminton.com/rankings/2/bwf-world-rankings/10/mixed-doubles/{year}/{week}?rows_per_page=5000'
```

```python
def getBWFRankings(year, week):
    headers = {'Accept-Encoding': 'deflate'}
    r = requests.get(xd_test.format(year=year, week=week), headers=headers)
    soup = BeautifulSoup(r.text, 'lxml')
    print soup.prettify()
    downs = soup.find_all('img', {'src': re.compile('down\.png')})
    for d in downs:
        d.parent.find('span', class_='ranking-change').string = '-' + d.parent.find('span', class_='ranking-change').string

    table = soup.find('table', class_='tblRankingLanding')
    table = pd.read_html(str(table))[0]
    return table
```
