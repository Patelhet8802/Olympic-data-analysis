```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_tracks = pd.read_csv('D:/olm/bwf_md_2015w1.csv')
df_tracks.head()
```

```python
pd.isnull(df_tracks).sum()
```

```python
df_tracks.describe().transpose()
```

```python
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
print(len(sample_df))
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, x = "PRIZE MONEY", y = "PLAYER", color = 'c').set(title = "winning prize")
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, x = "PRIZE MONEY", y = "PLAYER", color = 'b').set(title = "Popularity vs Acousticness")
```

```python
corr_df = df_tracks.drop(["PLAYER","PRIZE MONEY","RANK"], axis = 1).corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap = sns.heatmap(corr_df, annot = True,fmt = ".1g",vmin = -1, vmax = 1, center = 0, cmap = "inferno", linewidths=1, linecolor="Black")
heatmap.set_title("correlation between heatmap variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation = 90)
```
